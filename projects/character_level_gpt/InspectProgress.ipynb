```python
import torch
from tokenizer import CharacterLevelTokenizer
import pickle
with open("tokenizer.pk", "rb") as ftok:
    tokenizer = pickle.load(ftok)
```

```python
import random
gpt = torch.load("gpt.pt", weights_only=False)
device = next(gpt.parameters()).device
batch_size, seqlen = 8, 128
start = tokenizer.encode([random.choice(tokenizer.vocab) for _ in range(batch_size)])[:,None].to(device)
batches = gpt.batch_generate(start, seqlen)
for sample in tokenizer.decode(batches):
    print("########################\n",sample,"\n########################\n")
```
