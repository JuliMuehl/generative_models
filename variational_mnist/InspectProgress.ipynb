```python
import matplotlib.pyplot as plt
import torch
from itertools import product
%matplotlib inline
plt.gray()
```

```python
n = 10
fig, axs = plt.subplots(n,n)
vae = torch.load("mnist_vae.pt", weights_only=False)
samples = vae.sample(n**2).to("cpu").detach().numpy().swapaxes(1,-1).swapaxes(-2,-3)
samples = samples.reshape([n,n, *samples.shape[1:]])
for i, j in product(range(n),repeat=2):
    axs[i][j].axis("off")
    axs[i][j].imshow(samples[i,j])
```
